# Axiomas de la probabilidad

$$ P(A) = \frac{\text{número de resultados favorables}}{\text{número total de resultados posibles}} $$


```python
num_resultados_favorables = 4  # Hay 4 ases en una baraja
num_total_resultados = 52  # Total de cartas en la baraja

probabilidad = num_resultados_favorables / num_total_resultados
print("La probabilidad de sacar un as es:", probabilidad)
print("La probabilidad de sacar un as en %:", round(probabilidad * 100,2))
```
